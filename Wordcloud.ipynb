```
#importing the novel Message from Venus
f = open('message from venus.txt')
file_contents= f.read()
print(file_contents)
```

```
def calculate_frequencies(inp):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the","The","in", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just",\
     "copyright","about","for","if","You","may","There","But","into","on","not","said", "going"]
    
    non_pun_text=""
    for term in inp:
        if term not in punctuations:
            non_pun_text= non_pun_text + term
    words=non_pun_text.split()
    acceptable=[] #Words with meanings that you are useful
    frequencies={} #Number of times they appear
    for i in words:
        if i.isalpha():
            if i not in uninteresting_words:
              acceptable.append(i) 

    for alpha_word in acceptable:
        if alpha_word not in frequencies:
            frequencies[alpha_word]=1
        else:
            frequencies[alpha_word]+=1
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(frequencies)
    return cloud.to_array()
    
```

```
myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
